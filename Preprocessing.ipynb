```python
df3['Date_of_Birth']= pd.to_datetime(df3['Date_of_Birth']).dt.strftime("%Y-%m-%d")
df1['Date_of_Purchase']= pd.to_datetime(df1['Date_of_Purchase']).dt.strftime("%Y-%m-%d")
```

```python
for i in df3['Date_of_Birth']:
    y=i[0:4]
    t=date.today()
    if(t.year<int(y)):
        y=str(int(y)-100)
        x=y+i[4:]
        df3['Date_of_Birth']=df3['Date_of_Birth'].replace([i],x)
```

```python
for i in df1:
    df1[i].fillna(" ",inplace= True)

for i in df2:
    if (i=='phone_no') or (i=='Zipcode'):
        df2[i].fillna("0",inplace= True)
        continue
    df2[i].fillna(" ",inplace= True)

for i in df3:
    df3[i].fillna(" ",inplace= True)
```
